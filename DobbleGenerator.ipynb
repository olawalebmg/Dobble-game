```python
# Firstly,import emoji and store in dictionary.

import emoji
image_dict = dict()
fin = open('emoji_names.txt',"r")
lines = fin.readlines()
for i, el in enumerate(lines):
    image_dict[i+1] = emoji.emojize(el.strip())


# To create the deck and add the cards on dictionary
# where key is equal to card ID
# and value is equal to emoji ID

# we require 8 images per card

n = 7
r = range(n)
y = range(n+1)
card = 0

#Dictionary Initialization
deck = dict()

#First card
card += 1
# set to stote emoji Id
image_id = set()
for i in y:

#Append emoji Id to set
    image_id.add(i+1)

# Add card to dictionary
deck[card]=image_id
print()


#set to store emoji ID and add card 

# n cards 
image_id = set()
for j in r:
    card = card+1
    image_id.add(1)
    deck[card]=image_id
    
    image_id = set()
    for k in r:
        image_id.add(1)
        image_id.add(n+2 + n*j +k)
        deck[card]=image_id
        
# n x n cards
for i in r:
    image_id = set()
    for j in r:
        card = card+1
        image_id.add(i+2)
        deck[card]=image_id
        
        image_id = set()
        for k in r:
            image_id.add(i+2)
            image_id.add((n+1 +n*k + (i*k+j) % n)+1)
            deck[card]=image_id


# validity function to compare all the cards.
def validity(deck, flag = False):
   
    for card in deck:
        for other_card in deck:
# Here don't compare same cards
            if card != other_card:
# if flag = True,

                if flag:
                    print('check card n.', card, 'with .', other_card )

                    check = deck[card].intersection(deck[other_card])
                    print ('common ID ', check, 'there is ', len(check), ' common image')
                    

#error if set has length greater than 1 and more than 1 image is common
# exit loop
                    if len(check) > 1:
                        print ('incorrect deck')
                        return 

                else:
                    check = deck[card].intersection(deck[other_card])
                    if len(check) > 1:
                        print ('incorrect deck')
                        return 
                    

    if len(check) == 1:
        print('correct deck')

#  DobbleCard Class
#  __init__  : card ID to get the images ID
#  __repr__  : represent the card as requested


class DobbleCard():

    def __init__(self, cardID):
        self.imagesIDs = deck[cardID]
         
    def __repr__(self):

        self.images = ()
        for r in self.imagesIDs:
 
            self.images += (image_dict[r],)

        return ("\n".join(["".join(self.images[i:i+3]) for i in range(0,len(self.images),3)]))
            


# DobbleDeck Class 
# __init__ : Create deck with no. of cards requested, which calls the add_card
# add_card produces random key and it will add the card to respective deck
# remove_card will remove the played card from the deck
# play_card produces random index on the basis of deck size
# display card
# remove it from deck


import random

class DobbleDeck():

    def __init__ (self, c_number = 57):

        self.my_deck = []

        for card in range(0, c_number):

            self.add_card()
       
    def add_card(self):

        rand_key = random.randint(1, 57)

        self.my_deck += [DobbleCard(rand_key)]

    
    def remove_card(self, i):
# pop function : removes the card which is already played
        self.my_deck.pop(i)
        
    def play_card(self):

        r = random.randint(0, len(self.my_deck)-1)

        self.played_card = self.my_deck[r]

        if len(self.my_deck) > 1:
            self.remove_card(r)
        return self.played_card
         

def run_dobble():
    print('*WELCOME TO DOBBLE*')
    
# enter a number
    while True:
        try:
# how many cards to play?
            c_number = int(input('How many cards (<56)? '))
        except ValueError:
# enter a number ,then print error 
# enter input again
            print("That's not a correct number!")
        else: 
# exit loop 
            break
            
    while c_number >= 56 or c_number <=0:
        print("You have selected wrong number of card , Please try again")
        c_number = int(input('How many cards (<56) '))
    else:
        print("If you want to record a draw type \"d\" or \"D\"")
        deck = DobbleDeck(c_number+1)

        counter_one = 0
        counter_two = 0
# begin counter for game
        counter = 1
        
        l_card = str(deck.play_card())
        r_card = str(deck.play_card())

        print(l_card[0:3],'         ', r_card[0:3])
        print(l_card[4:7],'         ', r_card[4:7])
        print(l_card[8:10],'           ', r_card[8:10])
        
        winner = input('Who wins (A or B)?') 

        while winner not in ['a','A','b','B','d','D']:
            print('Sorry wrong input.Please try again!')
            winner = input('Who wins? (A or B)')
        else:

            if winner == 'a' or winner == 'A':
                counter_one += 1
            elif winner == 'b' or winner == 'B':
                counter_two += 1
            elif winner == 'd' or winner == 'D':
                print('That\'s a draw!')
            

            while counter < c_number: 

                l_card = r_card
 
                r_card = str(deck.play_card())

                print(l_card[0:3],'         ', r_card[0:3])
                print(l_card[4:7],'         ', r_card[4:7])
                print(l_card[8:10],'           ', r_card[8:10])
# increase game counter
                counter +=1
                winner = input('Who wins (A or B)?')

                while winner not in ['a','A','b','B','d','D']:
                    print('sorry wrong input. Please try again!')
                    winner = input('Who wins (A or B)?')
                else:
                    if winner == 'a' or winner == 'A':
                        counter_one += 1
                    elif winner == 'b' or winner == 'B':
                        counter_two += 1
                    elif winner == 'd' or winner == 'D':
                        print('That\'s a draw!')
            print() 
# print final score and winner
            print ('Score')
            print ('A:', counter_one)
            print ('B:', counter_two)
            print()
            if counter_one > counter_two:
                print ('Player A wins!')
            elif counter_two > counter_one:
                print ('Player B wins!')
            else:
                print ('That\'s a draw!')


run_dobble()
```
